have a look http://gmsh.info/ to install gmsh sdk

```python
import gmsh
import math
import numpy as np

lcmin=0.01
lcmax=0.1


# matrix
gmsh.initialize()
    
gmsh.option.setNumber("General.Terminal", 1)
#gmsh.option.setNumber("Mesh.MshFileVersion", 2.2)
gmsh.option.setNumber("Mesh.CharacteristicLengthMin", lcmin)
gmsh.option.setNumber("Mesh.CharacteristicLengthMax", lcmax)
#del quad
#gmsh.option.setNumber("Mesh.Algorithm",8)
#delaunay
#gmsh.option.setNumber("Mesh.Algorithm",5)
#meshadapt
gmsh.option.setNumber("Mesh.Algorithm",2)
#
gmsh.model.add('Matrice')
gmsh.model.setCurrent('Matrice')
#    
# contour
gmsh.model.geo.addPoint(-1, -1, 0,lcmax, tag=1)
gmsh.model.geo.addPoint(-0.2, -1, 0,lcmin, tag=2)
gmsh.model.geo.addPoint(-0.2, 0.4, 0,lcmin, tag=3)
gmsh.model.geo.addPoint(0.2, 0.4, 0,lcmin, tag=4)
gmsh.model.geo.addPoint(0.2, -1, 0,lcmin, tag=5)
gmsh.model.geo.addPoint(1., -1, 0,lcmax, tag=6)
gmsh.model.geo.addPoint(1, 1, 0,lcmax, tag=7)
gmsh.model.geo.addPoint(-1, 1, 0,lcmax, tag=8)
#
gmsh.model.geo.addLine(1, 2, tag=1)
gmsh.model.geo.addLine(2, 3, tag=2)
gmsh.model.geo.addLine(3, 4, tag=3)
gmsh.model.geo.addLine(4, 5, tag=4)
gmsh.model.geo.addLine(5, 6, tag=5)
gmsh.model.geo.addLine(6, 7, tag=6)
gmsh.model.geo.addLine(7, 8, tag=7)
gmsh.model.geo.addLine(8, 1, tag=8)
#
gmsh.model.geo.addCurveLoop([1, 2, 3, 4, 5, 6, 7, 8], tag=1)
#
#hole left
gmsh.model.geo.addPoint(-0.6, -0.8, 0,lcmin, tag=10)
gmsh.model.geo.addPoint(-0.7, -0.8, 0,lcmin, tag=11)
gmsh.model.geo.addPoint(-0.6, -0.7, 0,lcmin, tag=12)
gmsh.model.geo.addPoint(-0.5, -0.8, 0,lcmin, tag=13)
gmsh.model.geo.addPoint(-0.6, -0.9, 0,lcmin, tag=14)
#
gmsh.model.geo.addCircleArc(11, 10, 12, tag=10)
gmsh.model.geo.addCircleArc(12, 10, 13, tag=11)
gmsh.model.geo.addCircleArc(13, 10, 14, tag=12)
gmsh.model.geo.addCircleArc(14, 10, 11, tag=13)
#
gmsh.model.geo.addCurveLoop([10, 11, 12, 13], tag=10)
#
#hole right
gmsh.model.geo.addPoint( 0.6, -0.8, 0,lcmin, tag=20)
gmsh.model.geo.addPoint( 0.7, -0.8, 0,lcmin, tag=21)
gmsh.model.geo.addPoint( 0.6, -0.7, 0,lcmin, tag=22)
gmsh.model.geo.addPoint( 0.5, -0.8, 0,lcmin, tag=23)
gmsh.model.geo.addPoint( 0.6, -0.9, 0,lcmin, tag=24)
#
gmsh.model.geo.addCircleArc(21, 20, 22, tag=20)
gmsh.model.geo.addCircleArc(22, 20, 23, tag=21)
gmsh.model.geo.addCircleArc(23, 20, 24, tag=22)
gmsh.model.geo.addCircleArc(24, 20, 21, tag=23)
#
gmsh.model.geo.addCurveLoop([20, 21, 22, 23], tag=20)

gmsh.model.geo.addPlaneSurface([1,-10,-20], tag=1)
    
gmsh.model.geo.synchronize()
        
up = gmsh.model.addPhysicalGroup(1, [7], tag=1)
gmsh.model.setPhysicalName(1, up, 'up')

inside = gmsh.model.addPhysicalGroup(1, [2, 3, 4], tag=2)
gmsh.model.setPhysicalName(1, inside, 'inside')

matrix = gmsh.model.addPhysicalGroup(2, [1], tag=1)
gmsh.model.setPhysicalName(2, matrix, 'matrix')

#gmsh.fltk.run()
    
# Mesh generation and export
gmsh.model.mesh.setRecombine(2, 1)
gmsh.model.mesh.generate(2)
gmsh.write('matrix.msh')

gmsh.finalize()
#
# inclusion
gmsh.initialize()

gmsh.option.setNumber("General.Terminal", 1)
#gmsh.option.setNumber("Mesh.MshFileVersion", 2.2)
gmsh.option.setNumber("Mesh.CharacteristicLengthMin", lcmin)
gmsh.option.setNumber("Mesh.CharacteristicLengthMax", lcmax)
#
gmsh.option.setNumber("Mesh.Algorithm",2)
#
gmsh.model.add('Inclusion')
gmsh.model.setCurrent('Inclusion')
    
# contour
gmsh.model.geo.addPoint(-0.2, -1, 0,lcmax, tag=1)
gmsh.model.geo.addPoint(-0.2, 0.4, 0,lcmax, tag=2)
gmsh.model.geo.addPoint(0.2, 0.4, 0,lcmax, tag=3)
gmsh.model.geo.addPoint(0.2, -1, 0,lcmax, tag=4)
#
gmsh.model.geo.addLine(1, 2, tag=1)
gmsh.model.geo.addLine(2, 3, tag=2)
gmsh.model.geo.addLine(3, 4, tag=3)
gmsh.model.geo.addLine(4, 1, tag=4)
#
gmsh.model.geo.addCurveLoop([-1, -4, -3, -2], tag=1)
#
gmsh.model.geo.addPlaneSurface([1], tag=1)
#    
gmsh.model.geo.synchronize()
        
down = gmsh.model.addPhysicalGroup(1, [4], tag=1)
gmsh.model.setPhysicalName(1, down, 'down')

outside = gmsh.model.addPhysicalGroup(1, [1, 2, 3], tag=2)
gmsh.model.setPhysicalName(1, outside, 'outside')

inclusion = gmsh.model.addPhysicalGroup(2, [1], tag=1)
gmsh.model.setPhysicalName(2, inclusion, 'inclusion')

#gmsh.fltk.run()
    
# Mesh generation and export
gmsh.model.mesh.generate(2)
gmsh.write('inclusion.msh')

gmsh.finalize()
```

## Creating sample

```python
from pathlib import Path
import math

from pylmgc90 import pre

datbox = Path('./DATBOX')
datbox.mkdir(exist_ok=True)

dim = 2

# containers
ps = pre.avatars()
ms = pre.models()
mx = pre.materials()
svs   = pre.see_tables()
tacts = pre.tact_behavs()

# models
mt = pre.model(name='T3MLx', physics='MECAx', element='T3xxx', dimension=dim, external_model='MatL_',kinematic='large',
               formulation='TotaL',material='neoh_',anisotropy='iso__',mass_storage='lump_')
ms.addModel(mt)

mq = pre.model(name='Q4MLx', physics='MECAx', element='Q4xxx', dimension=dim, external_model='MatL_',kinematic='large',
               formulation='TotaL',material='neoh_',anisotropy='iso__',mass_storage='lump_')
ms.addModel(mq)

# materials
mat_matrix = pre.material(name='xxxxx',materialType='ELAS',elas='standard',
                          young=1.,nu=0.49,anisotropy='isotropic',
                          density=2e-9)
mx.addMaterial(mat_matrix)

mat_inclusion = pre.material(name='yyyyy',materialType='ELAS',elas='standard',
                             young=210000.,nu=0.3,anisotropy='isotropic',
                             density=7.8e-9)
mx.addMaterial(mat_inclusion)

# read matrix mesh  
mesh_matrix = pre.readMesh('matrix.msh',dim,keep_elements=[1])
# renumbering
mesh_matrix.rankRenumbering()

matrix = pre.avatar(dimension=dim)
# add bulks
matrix.addBulks(mesh_matrix.bulks)
# add nodes
matrix.addNodes(mesh_matrix.nodes)
# define groups
matrix.defineGroups()

# define model
if matrix.hasGroup('T3xxx'):
    matrix.defineModel(group='T3xxx', model=mt)
if matrix.hasGroup('Q4xxx'): 
    matrix.defineModel(group='Q4xxx', model=mq)

# define material
matrix.defineMaterial(material=mat_matrix)
#
matrix.addContactors(group='inside',shape='CLxxx',color='REDxx',weights=[0.25,0.75],reverse=True)
matrix.imposeDrivenDof(group='up',component=[2],dofty='vlocy')
ps.addAvatar(matrix)

# read inclusion
mesh_inclusion = pre.readMesh('inclusion.msh',dim)
# renumbering
mesh_inclusion.rankRenumbering()
#
inclusion = pre.avatar(dimension=dim)
# add bulks
inclusion.addBulks(mesh_inclusion.bulks)
# add nodes
inclusion.addNodes(mesh_inclusion.nodes)
# define groups
inclusion.defineGroups()
# define models
if inclusion.hasGroup('T3xxx'):
    inclusion.defineModel(group='T3xxx', model=mt)
if inclusion.hasGroup('Q4xxx'): 
    inclusion.defineModel(group='Q4xxx', model=mq)
# define material
inclusion.defineMaterial(material=mat_inclusion)
#
inclusion.addContactors(group='outside',shape='ALpxx',color='BLUEx')
# 
inclusion.imposeDrivenDof(group='down',component=2,dofty='vlocy',ct=-3.,rampi=0.,ramp=1e3)
inclusion.imposeDrivenDof(group='down',component=1,dofty='vlocy')
ps.addAvatar(inclusion)

# contact law
b = pre.tact_behav( name='expo_', law='EXPO_CZM',dyfr=0.30,stfr=0.30,
                    cn=1.e+12, s1=0.4, G1=75e-6,
                    ct=1.e+12, s2=0.4, G2=75e-6,
                    eta=1e-4)
tacts+=b

# visibility table
sv = pre.see_table(CorpsCandidat   ='MAILx', candidat   ='CLxxx', colorCandidat   ='REDxx',
                   CorpsAntagoniste='MAILx', antagoniste='ALpxx', colorAntagoniste='BLUEx',
                   behav=b, alert=.01, halo=1.)
svs+=sv

# postpro

post = pre.postpro_commands()
mecax_sets = [[(inclusion,"down")],[(matrix,"up")]]

post.addCommand(pre.postpro_command(name='NEW MECAx SETS', mecax_sets=mecax_sets))
post.addCommand(pre.postpro_command(name='Dep EVOLUTION', step=1))
post.addCommand(pre.postpro_command(name='Fint EVOLUTION', step=1))

# writtings
pre.writeDatbox(dim, mx, ms, ps, tacts, svs, post=post, datbox_path=datbox, gravy=[0., 0., 0.,])
try:
    pre.visuAvatars(ps)
except:
    pass
```

## Computation 

```python
import numpy as np 

# importing chipy module
from pylmgc90 import chipy

# Initializing
chipy.Initialize()

# checking/creating mandatory subfolders
chipy.checkDirectories()

# logMes
chipy.utilities_DisableLogMes()

#
# defining some variables
#

# space dimension
dim = 2

# modeling hypothesis ( 1 = plain strain, 2 = plain stress, 3 = axi-symmetry)
mhyp = 1

# time evolution parameters
dt = 1e-5
t_final = 2*dt #10.

dt_min = 1.e-6
dt_max = 1e-4

NR_max_iter = 20
NR_adapt = 5
NR_tol = 1.e-6

# theta integrator parameter
theta = 0.5

# interaction parameters
Rloc_tol = 5.e-2

chipy.CLxxx_SetNbNodesByCLxxx(2)
chipy.CLALp_Trim()

# nlgs parameters
tol    = 1.e-6
relax  = 1.0
norm   = 'Quad '
gs_it1 = 100
gs_it2 = 200
solver_type='Stored_Delassus_Loops         '

# write parameter
freq_write   = 1

# display parameters
freq_display = 1

#
# read and load
#

# Set space dimension
chipy.SetDimension(dim,mhyp)
#
chipy.utilities_logMes('INIT TIME STEPPING')
chipy.TimeEvolution_SetTimeStep(dt)
chipy.Integrator_InitTheta(theta)

# Newton loop parameters:
chipy.NewtonRaphson_SetFinalTime(t_final)
chipy.NewtonRaphson_SetMinTimeStep(dt_min)
chipy.NewtonRaphson_SetMaxTimeStep(dt_max)
chipy.NewtonRaphson_SetMaxIter(NR_max_iter)
chipy.NewtonRaphson_SetIncPatience(NR_adapt)

#
chipy.utilities_logMes('READ BEHAVIOURS')
chipy.ReadBehaviours()
chipy.ReadModels()
#
chipy.utilities_logMes('READ BODIES')
chipy.ReadBodies()
#
chipy.utilities_logMes('LOAD BEHAVIOURS')
chipy.LoadBehaviours()
chipy.LoadModels()
#
chipy.utilities_logMes('READ INI DOF')
chipy.ReadIniDof()
#
chipy.utilities_logMes('READ INI GPV')
chipy.ReadIniGPV()
#
chipy.utilities_logMes('READ DRIVEN DOF')
chipy.ReadDrivenDof()
#
chipy.utilities_logMes('LOAD TACTORS')
chipy.LoadTactors()
#
chipy.utilities_logMes('READ INI Vloc Rloc')
chipy.ReadIniVlocRloc()

#
# paranoid writes
#
chipy.utilities_logMes('WRITE BODIES')
chipy.WriteBodies()
chipy.utilities_logMes('WRITE BEHAVIOURS')
chipy.WriteBehaviours()
chipy.utilities_logMes('WRITE DRIVEN DOF')
chipy.WriteDrivenDof()

#
# open display & postpro
#
chipy.registerInterInternals('beta')
chipy.addRegistersToDisplay(True)

chipy.utilities_logMes('DISPLAY & WRITE')
chipy.OpenDisplayFiles()
chipy.OpenPostproFiles()

#
# simulation part ...
#
chipy.utilities_logMes('COMPUTE MASS')
chipy.ComputeMass()

while chipy.TimeEvolution_GetTime() < t_final :
    #
    chipy.utilities_logMes('INCREMENT STEP')
    chipy.IncrementStep()

    chipy.utilities_logMes('COMPUTE Fext')
    chipy.ComputeFext()

    # Newton loop
    chipy.NewtonRaphson_Initialize(NR_tol)
    is_converged = 1
    k=0
    #looping until something changes in CheckConvergence
    while is_converged == 1 :
        k+=1
        chipy.utilities_logMes('COMPUTE BULK')
        chipy.ComputeBulk()

        chipy.mecaMAILx_ComputeRayleighDamping(0.,0.01,[1])
    
        chipy.utilities_logMes('ASSEMB RHS/KT')
        chipy.AssembleMechanicalRHS()
        chipy.AssembleMechanicalLHS()

        chipy.utilities_logMes('COMPUTE Free Vlocy')
        chipy.ComputeFreeVelocity()
        #
        chipy.utilities_logMes('SELECT PROX TACTORS')
        chipy.SelectProxTactors()
        #
        ### Signorini Coulomb
        chipy.RecupRloc()
        chipy.ExSolver(solver_type, norm, tol, relax, gs_it1, gs_it2)
        chipy.StockRloc()
        ###
        chipy.utilities_logMes('COMPUTE DOF')
        chipy.ComputeDof()
        #
        if k > 1:
            NR_norm = chipy.mecaMAILx_ComputeResidueNorm()
            is_converged = chipy.NewtonRaphson_CheckConvergence(NR_norm)

    ### end while NR

    chipy.utilities_logMes('COMPUTE TIME STEP')
    #istate = 1 => redo step
    #istate = 2 => stop

    istate = chipy.NewtonRaphson_ComputeTimeStep()

    if not istate == 1 :

        chipy.utilities_logMes('UPDATE TACT BEHAV')
        chipy.UpdateTactBehav()
        chipy.StockRloc()

        chipy.utilities_logMes('UPDATE DOF')
        chipy.UpdateStep()
        #
        ### write results ###
        #
        chipy.WriteOutDof(freq_write)

        chipy.WriteOutGPV(freq_write)

        chipy.WriteOutVlocRloc(freq_write)
        chipy.WriteDisplayFiles(freq_display)

        chipy.WritePostproFiles()

        if istate == 2 :
            # istate => Stop
            break

### end while time loop ###

#
# close display & postpro
#
chipy.CloseDisplayFiles()
chipy.ClosePostproFiles()

# this is the end
chipy.Finalize()
```
